```python
# Replace with the most frequent value (Mode)
df['Engagement_with_Ads'] = df['Engagement_with_Ads'].fillna(df['Engagement_with_Ads'].mode()[0])
df['Social_Media_Influence'] = df['Social_Media_Influence'].fillna(df['Social_Media_Influence'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
# To create an age_group column
bins = [0, 18, 35, 60, 100]
labels = ['Young', 'Adult', 'Middle Aged', 'Senior']


df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

```

```python
print(df[['age', 'age_group']].head(10))
```

```python
# To create a column called purchase efficiency
# 1. Remove '$' and ',' then convert to float
df['purchase_amount'] = df['purchase_amount'].replace('[\$,]', '', regex=True).astype(float)

# 2. Now run the division
df['purchase_efficiency'] = df['purchase_amount'] / df['time_to_decision']

# 3. View the result
print(df[['purchase_amount', 'time_to_decision', 'purchase_efficiency']].head())
```

```python
# Marks users who are members AND gave a rating of 8 or higher
df['is_vip'] = (df['customer_loyalty_program_member'] == True) & (df['product_rating'] >= 8)
```

```python
# Mapping text to numbers to create a combined score
influence_map = {'Low': 1, 'Middle': 2, 'High': 3}
df['digital_score'] = df['social_media_influence'].map(influence_map) + df['engagement_with_ads'].map(influence_map)
```
